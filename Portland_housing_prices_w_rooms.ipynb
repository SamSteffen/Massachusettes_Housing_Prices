```python
# import dependencies
import pandas as pd
import os
import csv
from datetime import datetime
import datetime as dt
import time
```

```python
# read in the data as a df
file_path = 'C:/Users/ssteffen/desktop/sam/Vanderbilt Boot Camp/MyRepo/Module 20_Final/Final Project/Resources\portland_housing.csv'

raw_df = pd.read_csv(file_path)
raw_df.head()
```

```python
# Explore data columns
columns = raw_df.columns.tolist()
columns
```

```python
#call a new df with only relevant data columns
df = raw_df[[
     'abbreviatedAddress',
     'address/city',
     'address/zipcode',
     'bathrooms',
     'bedrooms',
     'dateSold',
     'daysOnZillow',
     'latitude',
     'longitude',
     'lotSize',
     'price',
     'propertyTaxRate',
     'rentZestimate',
     'taxAssessedValue',
     'taxAssessedYear',
     'zestimate',
]]

df
```

```python
#drop NaN values
df_clean = df.dropna(how="any")
df_clean
```

```python
df_clean.dtypes
```

```python
# convert the date type columns to datetype data in new columns, then drop the old columns
df_clean['Date_sold'] = pd.to_datetime(df_clean['dateSold'], unit='ms').dt.date
df_clean['bathrooms'] = df_clean['bathrooms'].astype(int) 
df_clean['bedrooms'] = df_clean['bedrooms'].astype(int)
df_clean['lotSize'] = df_clean['lotSize'].astype(int)
df_clean['daysOnZillow'] = df_clean['daysOnZillow'].astype(int)
df_clean['taxAssessedYear'] = df_clean['taxAssessedYear'].astype(int)

df_clean = df_clean.drop(columns=[
    'dateSold'])

df_clean
```

```python
# reorder the columns in the desired output format
# df_clean.columns
df_clean = df_clean[[
    'abbreviatedAddress',
    'address/city', 
    'address/zipcode', 
    'latitude', 
    'longitude',
    'bathrooms',
    'bedrooms',
    'lotSize',
    'daysOnZillow',
    'Date_sold',
    'price', 
    'zestimate', 
    'rentZestimate',
    'propertyTaxRate', 
    'taxAssessedValue', 
    'taxAssessedYear', 
]]

df_clean.head()
```

```python
#drop duplicate rows, if any
df_clean = df_clean.drop_duplicates()
df_clean.count()
```

```python
# sort the data by date sold, beginning with the most recent
df_clean = df_clean.sort_values(by='Date_sold', ascending=False)
df_clean = df_clean.reset_index(drop=True)
df_clean
```

```python
# find the bottom of the date range of the dataset
df_clean['Date_sold'].min()
```
